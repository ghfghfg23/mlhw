```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'figure.max_open_warning': 0})
```

```python
benign_df = df.loc[df['diagnosis'] == 'B']
malignant_df = df.loc[df['diagnosis'] == 'M']
def plot_hist(column):
    sns.distplot(benign_df[column], label='B', color='blue', kde=False, norm_hist=False)
    sns.distplot(malignant_df[column], label='M', color='red', kde=False, norm_hist=False)
    
def plot_hist_grid():
    features = sorted(list(df.drop(columns=['diagnosis'])))
    feature_count = len(features)
    fig, axes = plt.subplots(nrows=10, ncols=3, figsize=(15, 30))
    fig.subplots_adjust(hspace=0.5)
    for ax, name in zip(axes.flatten(), features):
        sns.distplot(benign_df[name], ax=ax, label='B', color='blue', kde=False, norm_hist=False)
        sns.distplot(malignant_df[name], ax=ax, label='M', color='red', kde=False, norm_hist=False)
```

```python
plot_hist_grid()
```

На гистограммах показаны распределения фич с разным диагнозом. Синие это доброкачественные, красные это злокачественные опухоли.
На большинстве гистограмм прослеживается зависимость что чем больше значение фичи, тем вероятнее что опухоль злокачественная. Наименее различима разница в таких фичах как texture, symmetry, smoothness, fractal_dimension.

```python
plt.figure(figsize=(32,20));
sns.heatmap(df.corr(), annot=True, fmt='.2f', vmin=-1.0, vmax=1.0, cmap=sns.diverging_palette(240, 10, n=100));
```

Наблюдается сильная корреляция между radius_mean, perimeter_mean, area_mean, radius_worst, perimeter_worst, area_worst (что логично, потому что это все геометрические взаимосвязные значения). Так же есть сильная корреляция между radius_se, texture_se, perimeter_se.
Средняя корелляция есть у compactness, concavity, concave_points почти со всеми другими фичами.
Отрицательная корелляция наблюдается только у fractal_dimension_mean и smoothness_se с геометрическими признаками, но довольно слабая (~0.25). 


```python
def plot_scatter(between_columns):
    sns.pairplot(df.filter(regex="|".join(between_columns)))
```

```python
plot_scatter(['radius', 'perimeter', 'area'])
```

Прослеживается линейная зависимость между радиусом, периметром и площадью.

```python
plot_scatter(['compactness', 'concavity', 'concave_points'])
```

Между compactness, concavity, concave_points зависимость тоже наблюдается, но разброс намного больше чем у геометрических параметров.

```python
plot_scatter(['fractal_dimension_mean', 'radius_mean'])
```

Тут вроде как должна быть обратная небольшая корелляция, но я ее не вижу.

```python
for column in ['area', 'compactness', 'concave points', 'concavity', 'fractal_dimension', 'perimeter', 'radius', 'smoothness', 'symmetry', 'texture']:
    plot_scatter([column])
```

Графики зависимостей между *_mean, *_se, *_worst. Все эти фичи между собой кореллируют. Наибольший разброс наблюдается в texture, symmetry и compactness

```python
def plot_box():
    features = sorted(list(df.drop(columns=['diagnosis'])))
    feature_count = len(features)
    fig, axes = plt.subplots(nrows=10, ncols=3, figsize=(15, 30))
    fig.subplots_adjust(hspace=0.5)
    for ax, name in zip(axes.flatten(), features):
        sns.boxplot(y="diagnosis", ax=ax, x=name, data=df)
        ax.artists[0].set_facecolor('red')
        ax.artists[1].set_facecolor('blue')

```

```python
plot_box()
```

Исходя из boxplot графиков делаю вывод что наиболее удобно определять диагноз было бы по area, concave_points_mean, concave_points_worst, perimeter, radius_mean, radius_worst т.к. у них минимальное пересечение коробок и не очень много выбросов
