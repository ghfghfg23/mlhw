```python
def non_correlated_columns(bound):
    corr = df.corr()
    corr = corr - np.identity(len(corr.columns))
    diagnosis_column = df['diagnosis'].map(lambda d: 1 if d == 'M' else 0).astype(float)
    result = []
    filter_list = []
    diag_corr = {}
    for column in corr.columns:
        correlated_columns = set(corr[column].loc[corr[column]>=bound].index.tolist())
        if len(correlated_columns) > 0:
            correlated_columns.add(column)
            if all(map(lambda added: len(added.symmetric_difference(correlated_columns)) > 0, filter_list)):
                filter_list.append(correlated_columns)
        else:
            result.append(column)
        diag_corr[column] = diagnosis_column.corr(df[column])

    filter_list = list(map(lambda f: list(f), filter_list))
    def max_index(data):
        idx = data.index(max(data))
        return idx
        
    def max_value(func, data):
        val = data[max_index(list(map(func, data)))]
        return val

    while filter_list:
        maximums = list(map(lambda f: max_value(lambda c: abs(diag_corr[c]), list(f)), list(filter_list)))
        d = list(map(lambda f: abs(diag_corr[f]), maximums))
        maximum = max_index(d)
        added = maximums[maximum]
        result.append(added)
        removed = filter_list[maximum]
        removed.remove(added)
        def safe_remove(data, value):                
            try:
                data.remove(value) 
            except ValueError:
                pass
        filter_list.pop(maximum)
        for f in filter_list:
            if f.count(added) > 0:
                for r in f:
                    if r != added and not removed.count(r):
                        removed.append(r)
            safe_remove(f, added)
        for f in filter_list:
            for r in removed:
                safe_remove(f, r)
        filter_list = list(filter(lambda f: len(f) > 0, filter_list))
    return result
    
```

Суть алгоритма в следующем:
1. для каждого столбца выбираем все столбцы с корреляцией выше границы
2. вычисляем для каждого столбца корреляцию с целевой переменной, для того чтобы выбрать те столбцы которые максимально коррелируют с целевой переменной
3. нахожу из всех столбцов максимум корреляции с целевой переменной. после этого выкидываю все столбцы с которыми коррелирует найденный максимум
4. повторяю процедуру пока не отфильтрую все лишние

# ML

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
def quality_report(prediction, actual):
    print("Accuracy: {:.3f}\nPrecision: {:.3f}\nRecall: {:.3f}\nf1_score: {:.3f}".format(
        accuracy_score(prediction, actual),
        precision_score(prediction, actual),
        recall_score(prediction, actual),
        f1_score(prediction, actual)
    ))
```

```python
x = df[columns]
y = df['diagnosis'].map(lambda d: 1 if d == 'M' else 0).astype(float)
scaler = StandardScaler()
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
def plot_roc_auc(reg):
    prediction = reg.predict_proba(X_test_scaled)[:,1]
    fpr, tpr, thresholds = roc_curve(y_test, prediction)
    auc_score = roc_auc_score(y_test, prediction)
    plt.plot(fpr, tpr, label='ROC curve ')
    plt.plot([0, 1], [0, 1])
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC AUC: {:.3f}'.format(auc_score))
    plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=40)

knn.fit(X_train_scaled, y_train)
print("Train quality:")
quality_report(knn.predict(X_train_scaled), y_train)
print("\nTest quality:")
quality_report(knn.predict(X_test_scaled), y_test)
plot_roc_auc(knn)
```

```python
log_reg = LogisticRegression()

log_reg.fit(X_train_scaled, y_train)
print("Train quality:")
quality_report(log_reg.predict(X_train_scaled), y_train)
print("\nTest quality:")
quality_report(log_reg.predict(X_test_scaled), y_test)
plot_roc_auc(log_reg)
```

Логистическая регрессия из коробки в среднем дает результат лучше чем knn классификатор. У knn сильно хуже precision, но лучше recall. Учитывая специфику датасета, я полагаю что лучше использовать knn который имеет лучший recall чтобы не пропускать возможно больных людей.

```python
featureImportance = pd.DataFrame({"feature": columns, 
                                  "importance": log_reg.coef_[0]})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].plot(kind='bar');
```

Наиболее важными фичами стали radius_worst, radius_se, concave_points_worst, texture_worst. Все эти фичи были хорошо изображены на скаттерплотах, кроме texture_worst.

```python
log_reg_cv = LogisticRegressionCV(Cs=range(1, 300), cv=10)

log_reg_cv.fit(X_train_scaled, y_train)
print("Train quality:")
quality_report(log_reg_cv.predict(X_train_scaled), y_train)
print("\nTest quality:")
quality_report(log_reg_cv.predict(X_test_scaled), y_test)
plot_roc_auc(log_reg_cv)
print("Best C: ", log_reg_cv.C_)
```

```python
featureImportance = pd.DataFrame({"feature": columns, 
                                  "importance": log_reg_cv.coef_[0]})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].plot(kind='bar');
```

Качество модели с регуляризацией не изменилось, но изменились коэффициенты важности фич.

```python
plot_scatter(featureImportance.loc[abs(featureImportance['importance']) > 1.5].index)
```

Самые важные фичи оказались фичами которые лучше всего разделимы на графиках.
