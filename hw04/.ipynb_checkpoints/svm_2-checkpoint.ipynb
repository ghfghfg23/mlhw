```python
def number_eda(column):
    fig, ax =plt.subplots(2,2)
    sns.kdeplot(data[column], ax=ax[0,0], bw=1)
    for cat in target_categories:
        sns.kdeplot(data[data[target] == cat][column], ax=ax[0,1], bw=1)
    sns.boxplot(data[column], ax=ax[1,0])
    sns.boxplot(y=target, x=column, data=data, ax=ax[1,1])
    fig.show()
    display(Markdown('## {}'.format(column)))
    value_counts = data.groupby(target)[column].value_counts()
    table = pd.DataFrame()
    hist_bins = np.histogram(data[column])[1]
    for cat in target_categories:
        table[cat] = np.histogram(data[data[target] == cat][column], bins=hist_bins)[0]
    table['total'] = table[target_categories].sum(axis=1)
    table.index = pd.Series(hist_bins[:-1])
    display(table)
    plt.show()

def object_eda(column):
    fig, ax =plt.subplots(1,2)
    sns.countplot(x=column, data=data, ax=ax[0])
    sns.countplot(x=column, data=data, hue=target, ax=ax[1])
    fig.show()
    display(Markdown('## {}'.format(column)))
    value_counts = data.groupby(target)[column].value_counts()
    table = value_counts.unstack(level=0)
    table = table.fillna(0)
    table['total'] = table[target_categories].sum(axis=1)
    table.index.name = None
    table.columns.name = None
    table.sort_values(by=['total'], ascending=False, inplace=True)
    display(table)
    plt.show()
    
def feature_eda(column):
    if (pd.api.types.is_numeric_dtype(data[column].dtype)):
        number_eda(column)
    else:
        object_eda(column)
    

```

```python
def data_split_train_test(X, y, test_size=0.3, train_size=None):
    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=test_size,
        train_size=train_size,
        stratify=y,
        random_state=0)
    return X_train, X_test, y_train, y_test

def quality_report(prediction, actual):
    report_str = "Accuracy: \t {:.3f}\n" +\
                 "Precision: \t {:.3f}\n" +\
                 "Recall: \t {:.3f}\n" +\
                 "f1_score: \t {:.3f}\n" +\
                 "ROC-AUC: \t {:.3f}"
    
    acc = accuracy_score(prediction, actual)
    precision = precision_score(prediction, actual)
    recall = recall_score(prediction, actual)
    f1 = f1_score(prediction, actual)
    rocauc = roc_auc_score(prediction, actual)
    quality_list = [acc, precision, recall, f1, rocauc]
    
    return quality_list

quality = pd.DataFrame(
    [],
    columns=["Accuracy", "Precision", "Recall", "F1", "ROC-AUC"],
)

def make_examine(X, y, X_preproc, y_prep, train_size=None):
    pX = X_preproc.transform(X)
    py = y_prep.transform(y)
    def examine(tag, X_make_prep, model):
        X_train, X_test, y_train, y_test = data_split_train_test(pX, py, train_size=train_size)
        X_prep = X_make_prep()
        X_p = X_prep.fit_transform(X_train)
        model.fit(X_p, y_train)
        pred = model.predict(X_prep.transform(X_test))
        quality.loc[tag, :] = quality_report(pred, y_test)
    return examine
```

# Feature engeneering

```python
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
from sklearn.pipeline import make_union
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils.validation import check_is_fitted

class Transform(FunctionTransformer):
    def __init__(self, old, new, f):
        self.old = old
        self.new = new
        self.f = f
        super().__init__(self.forward_transform)
        
    def forward_transform(self, data):
        df = data.copy()
        t = self.f(df[self.old])
        t.index = df[self.old].index
        df[self.new] = t
        return df.drop(self.old, axis=1)

    def get_feature_names(self):
        return self.new

class LogTransform(FunctionTransformer):
    def __init__(self, columns):
        self.columns = columns
        super().__init__(self.forward_transform)
                         
    def forward_transform(self, data):
        df = data.copy()
        df[self.columns] = df[self.columns].apply(np.log1p)
        return df        
                             
    def get_feature_names(self):
        return self.columns
        
class ValueMapper(FunctionTransformer):
    def __init__(self, columns, mapping):
        self.columns = columns
        self.mapping = mapping
        super().__init__(self.replace)
        
    def replace(self, data):
        return data.replace({ x: self.mapping for x in self.columns})
        
    def get_feature_names(self):
        return self.columns
    
class Filter(TransformerMixin, BaseEstimator):
    def __init__(self, columns, pipeline):
        self.columns = columns
        self.pipeline = pipeline
    def fit(self, X, y=None):
        try:
            check_is_fitted(self.pipeline)
        except:
            self.pipeline = self.pipeline.fit(X[self.columns], y)
        return self
    def transform(self, X):
        X_t = self.pipeline.transform(X[self.columns])
        return X_t
    def fit_transform(self, X, y=None):
        self.fit(X, y)
        X_t = pd.DataFrame(self.transform(X[self.columns]), columns=self.get_feature_names(), index=X.index)
        return X_t
    def get_feature_names(self):
        try:
            return self.pipeline.get_feature_names()
        except:
            return self.columns

class Drop(Filter):
    def __init__(self, drop_columns):
        self.drop_columns = drop_columns
    def fit(self, X, y=None):
        self.columns = list(filter(lambda c: c not in self.drop_columns, X.columns.values.tolist()))
        self.pipeline = FunctionTransformer(lambda a: a)
        return self

class ObjectEncoder(Filter):
    def __init__(self, pipeline):
        self.pipeline = pipeline
    def fit(self, X, y=None):
        self.columns = X.select_dtypes([object]).columns.values.tolist()
        super().fit(X, y)
        return self
    
class NumberEncoder(Filter):
    def __init__(self, pipeline):
        self.pipeline = pipeline
    def fit(self, X, y=None):
        self.columns = X.select_dtypes([np.number]).columns.values.tolist()
        super().fit(X, y)
        return self
    
class BoolEncoder(Filter):
    def __init__(self, pipeline):
        self.pipeline = pipeline
    def fit(self, X, y=None):
        self.columns=X.select_dtypes([bool]).columns.values.tolist()
        super().fit(X, y)
        return self
    def transform(self, X):
        X_t = super().transform(X).astype(np.float)
        return X_t
    
class Empty(TransformerMixin, BaseEstimator):
    def fit(self, X, y=None):
        self.columns = X.columns
        return self
    def transform(self, X):
        return X
    def fit_transform(self, X, y=None):
        return X
    def get_feature_names(self):
        return self.columns
            
        
class Cyclic(FunctionTransformer):
    def __init__(self, min, max, column):
        self.column = column
        self.scaler = MinMaxScaler(feature_range=(min, max))
        super().__init__(self.forward_transform)
        
    def forward_transform(self, data):
        df = data.copy()
        df['sin_' + self.column] = np.sin(2 * np.pi * df[self.column])
        df['cos_' + self.column] = np.cos(2 * np.pi * df[self.column])
        return df.drop(self.column, axis=1)
    
    def get_feature_names(self):
        return ['sin_' + self.column, 'cos_' + self.column]            

class Merge(FunctionTransformer):
    def __init__(self, pipelines):
        self.pipelines = pipelines
        super().__init__(self.perform)
    def perform(self, data):
        ds = [pd.DataFrame(p.fit_transform(data), columns=p.get_feature_names()) for p in self.pipelines]
        d = pd.concat(ds, axis=1)
        return d
    def get_feature_names(self):
        return [name for names in map(lambda p: p.get_feature_names(), self.pipelines) for name in names]

examine = make_examine(
    X, 
    y, 
    Empty(),
    target_encoder,
    train_size=0.2
)
```

# Baseline

```python
from sklearn.linear_model import LogisticRegression
naive_encoding = lambda: Pipeline([
        ('encoding', Merge([
            ObjectEncoder(OneHotEncoder(sparse=False, handle_unknown='ignore')),
            NumberEncoder(StandardScaler()),
            BoolEncoder(Empty())
        ]))])

examine('baseline (logistic regression)', naive_encoding, LogisticRegression())
quality
```

```python
data['native-country'].value_counts().index.map(lambda x: x.strip()).map(lambda x: x in countries['name'])
mapping = pd.DataFrame(data['native-country'].value_counts().index, columns=['native-country'])
mapping['trimmed'] = mapping['native-country'].apply(lambda x: x.strip().replace('-', ' '))
mapping['has_name'] = mapping['trimmed'].isin(countries['name'])
mapping['name'] = mapping[mapping['has_name']]['trimmed']
mapping[mapping['has_name'] == False]
```

```python
def rename(trimmed, new):
    mapping.loc[mapping['trimmed'] == trimmed, 'name'] = new
rename('England', 'United Kingdom')
rename('Columbia', 'Colombia')
rename('Hong', 'Hong Kong')
rename('Trinadad&Tobago', 'Trinidad and Tobago')
rename('Yugoslavia', 'Serbia')
rename('Outlying US(Guam USVI etc)', 'Guam')
rename('Scotland', 'United Kingdom')
rename('Holand Netherlands', 'Netherlands')
rename('South', 'South Korea')
mapping['from-usa'] = mapping['native-country'] == ' United-States'
rename('?', 'United States')
country_mapping = pd.merge(mapping, countries, 'left', on='name')
```

```python
coordinates_columns = ['latitude', 'longitude', 'from-usa']
```

```python
examine_coordinates = make_examine(
    X,
    y,
    Pipeline([
        ('coordinates', Transform('native-country', 
                                  coordinates_columns, 
                                  lambda data: pd.merge(data, country_mapping[['from-usa', 'latitude', 'longitude', 'native-country']], 'left', on='native-country')[coordinates_columns])),
        ('cyclic latitude scaling', Cyclic(-90, 90, 'latitude')),
        ('cyclic longitude scaling', Cyclic(-180, 180, 'longitude'))
    ]),
    target_encoder,
    train_size=0.2
)
[examine_coordinates('svc with coordinates ({} kernel)'.format(kernel), naive_encoding, SVC(kernel=kernel)) for kernel in tqdm(['linear', 'poly', 'rbf', 'sigmoid'])]
quality
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import GridSearchCV

def X_y():
    targeted_data = data.copy().drop([target], axis=1)
    return (model(FunctionTransformer(lambda x: x)).fit_transform(targeted_data),
            data[target].copy().apply(lambda value: 1 if value == target_categories[1] else 0))
    
def search_params(kernel):
    X, y = X_y()
    C_range = np.logspace(-2, 10, 13)
    gamma_range = np.logspace(-9, 3, 13)
    param_grid = dict(gamma=gamma_range, C=C_range)
    cv = ShuffleSplit(n_splits=3, test_size=0.2, random_state=42)
    grid = GridSearchCV(SVC(kernel=kernel), param_grid=param_grid, scoring='accuracy', cv=cv, verbose=5, n_jobs=8)
    grid.fit(X, y)
    return grid
linear_kernel = search_params('linear')
```
